# Загрузка библиотек

```python
import numpy as np
import pandas as pd

# Decision Tree Classifier
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
# Перевод в категории
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics  import recall_score

from joblib import dump, load
```

# Классификация с помощью DecisionTreeClassifier()

```python
X_train = pd.read_csv('/content/drive/MyDrive/Базы/Metro/Datasets/datasets_df_clf/X_train_clf_df.csv')
X_test = pd.read_csv('/content/drive/MyDrive/Базы/Metro/Datasets/datasets_df_clf/X_test_clf_df.csv')
y_train = pd.read_csv('/content/drive/MyDrive/Базы/Metro/Datasets/datasets_df_clf/Y_train_clf_df.csv')
y_test = pd.read_csv('/content/drive/MyDrive/Базы/Metro/Datasets/datasets_df_clf/Y_test_clf_df.csv')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# fit a CART model to the data
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print(model)
# make predictions
expected = y_test
predicted = model.predict(X_test)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
# Predict with the best model.
y_pred_clf = model.predict(X_test)
```

```python
recall_metric = recall_score(expected, predicted, average='macro')
recall_metric
```

# Сохранение модели и проверка результата recall = 0,99

```python
# Сохраним файл с моделью
dump(model, '/content/drive/MyDrive/Базы/Metro/Models/DecisionTreeClassifier_v2.joblib')
clf = load('/content/drive/MyDrive/Базы/Metro/Models/DecisionTreeClassifier_v2.joblib')
```

```python
# Predict with the best model.
y_pred_clf = clf.predict(X_test)
```

```python
recall_metric = recall_score(expected, y_pred_clf, average='macro')
recall_metric
```
